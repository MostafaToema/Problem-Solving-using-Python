```python
import networkx as nx
```

```python
graph = {"A" : ["C", "D"],
         "B" : ["D", "E"],
         "C" : ["A", "D"],
         "D" : ["A", "B", "C", "E", "F"],
         "E" : ["B", "D", "F"],
         "F" : ["D", "E"]}
```

```python
graph
```

## 1) Draw Graph

```python
G = nx.Graph(graph)
nx.draw(G, with_labels = True, font_size=10)
```

## 2) DFS path for the Graph

```python
DFSGraph = nx.dfs_tree(G, source='A')
nx.draw(DFSGraph, with_labels=True)
print(list(DFSGraph))
```

## 3) BFS Path for the graph

```python
BFSGraph = nx.bfs_tree(G, source='A')
nx.draw(BFSGraph, with_labels=True)
print(list(BFSGraph))
```

## 4) Dijkstra

```python
nx.single_source_dijkstra_path(G, source='A')
```

## 5) Prim 

```python
primGraph = nx.minimum_spanning_tree(G, "A", algorithm="prim")
nx.draw(primGraph, with_labels=True)
```

## 6) Kruskal

```python
kruskalGraph = nx.minimum_spanning_tree(G, algorithm="kruskal")
nx.draw(kruskalGraph, with_labels=True)
```
